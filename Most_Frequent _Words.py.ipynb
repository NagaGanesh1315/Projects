#Problem 1:
Compute the Number of Times a Pattern Appears in a Text

```python
#Task 1


def Count(Text, Pattern):
  count=0

  for i in range(len(Text)-len(Pattern)+1):

    if Text[i:i+len(Pattern)]==Pattern:
     count+=1
  return count
    


print(Count('ACAACTATGCATACTATCGGGAACTATCCT','ACTAT'))
print(Count('CGATATATCCATAG','ATA'))
```

#Problem 2:
Find the Most Frequent Words in a String

```python
#Task 1

def mostFrequesntKmers(Text,K):

  freq=dict()

  for i in range(len(Text)-K+1):

    if Text[i:i+K] in freq:
      freq[Text[i:i+K]]+=1
    else:
      freq[Text[i:i+K]]=1
  
  #print(freq)
  maxiFreq=(max(freq.values()))
  frequentKmers=[]
  for i in freq:
    if freq[i]==maxiFreq:
      frequentKmers.append(i)
  return frequentKmers

mostFrequesntKmers('ACGTTGCATGTCGCATGATGCATGAGAGCT',4)
```

#Problem 3:
Find the Reverse Complement of a String

```python
#Task 1

def ReverseComplementDNA(Pattern):

  ComplementPattern=""
  for i in Pattern:

    if i=='A':
      ComplementPattern+='T'
    
    elif i=='T':
      ComplementPattern+='A'

    elif i=='C':
      ComplementPattern+='G'
    
    elif i=='G':
      ComplementPattern+='C'
  ReverseComplementPattern=ComplementPattern[::-1]
  return ReverseComplementPattern

print(ReverseComplementDNA('AAAACCCGGT'))
```

#Problem 4:
Find All Occurrences of a Pattern in a String

```python
# Task 1

def allOccurances(Text, Pattern):
  indices=[]

  for i in range(len(Text)-len(Pattern)+1):

    if Text[i:i+len(Pattern)]==Pattern:
     indices.append(i)
  return indices
    


print(allOccurances('ACAACTATGCATACTATCGGGAACTATCCT','ACTAT'))
print(allOccurances('GATATATGCATATACTT','ATAT'))
```
